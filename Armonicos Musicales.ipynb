```python
import math
# Definir la frecuencia de La (La natural)
la_frequency = 440  # Hz

# Definir el factor de multiplicación para calcular las otras notas
semitone_ratio = 2 ** (1/12)

# Lista de nombres de las notas
note_names = ["La", "La#", "Si", "Do", "Do#", "Re", "Re#", "Mi", "Fa", "Fa#", "Sol", "Sol#","La2"]

# Inicializar un array para almacenar las frecuencias de las notas
frequencies = np.zeros(len(note_names))

# Calcular y mostrar las frecuencias de las notas
print("Frecuencias de las notas musicales:")
for i, note_name in enumerate(note_names):
    frequencies[i] = la_frequency * (semitone_ratio ** i)
    print(f"{note_name}: {frequencies[i]:.2f} Hz")

# Graficar las frecuencias de las notas
plt.figure(figsize=(10, 4))
plt.plot(note_names, frequencies, marker='o', linestyle='-')
plt.title('Frecuencias de las notas musicales')
plt.xlabel('Nota')
plt.ylabel('Frecuencia (Hz)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Audio, display

# Definir la frecuencia fundamental de La (La natural)
la_frequency = 440  # Hz

armonico = 2 ** (1/12)

# Calcular la frecuencia del primer armónico (La una octava por encima)
la_harmonic_frequency = la_frequency * armonico
print(la_harmonic_frequency)

# Definir el tiempo
duration = 0.5  # en segundos
sample_rate = 48000  # tasa de muestreo (muestras por segundo)

# Crear un array de tiempo
t = np.linspace(0, duration, int(sample_rate * duration), endpoint=False)

# Calcular las amplitudes de las señales para La y su armónico
la_amplitude = 0.5
la_harmonic_amplitude = 0.5

# Definir el desfase en grados
harmonic_phase_shift_degrees = 0  # grados

# Convertir el desfase de grados a radianes
harmonic_phase_shift = np.deg2rad(harmonic_phase_shift_degrees)

# Generar las señales La y su armónico
la_signal = la_amplitude * np.sin(2 * np.pi * la_frequency * t)
la_harmonic_signal = la_harmonic_amplitude * np.sin(2 * np.pi * la_harmonic_frequency * t + harmonic_phase_shift)

# Sumar las señales para obtener la combinación de ambas
combined_signal = la_signal + la_harmonic_signal

plots = 4
# Graficar las señales
plt.figure(figsize=(12, 10))

plt.subplot(plots, 1, 1)
plt.plot(t, la_signal)
plt.title('La (440 Hz)')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True)

plt.subplot(plots, 1, 2)
plt.plot(t, la_harmonic_signal)
plt.title('Armónico de La (880 Hz)')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True)

plt.subplot(plots, 1, 3)
plt.plot(t, combined_signal)
plt.title('La y su armónico combinados')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True)

plt.subplot(plots, 1, 4)
plt.plot(t, la_signal)
plt.title('La (440 Hz)')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True)

plt.plot(t, la_harmonic_signal)
plt.title('Armónico de La (880 Hz)')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True)

plt.tight_layout()
plt.show()
display(Audio(data=combined_signal, rate=48000))
```
